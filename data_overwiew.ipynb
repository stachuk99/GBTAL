```python
import os
import pandas as pd
import numpy as np
import sklearn
import multi_imbalance
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.preprocessing import OneHotEncoder
from multi_imbalance.utils.data import load_arff_dataset
```

```python
dir = os.getcwd()
datasets_to_compare = ['glass', 'flare', 'new_ecoli', 'new_winequality-red', 'new_yeast', 'balance-scale']
```

```python
for dataset_name in datasets_to_compare:
    path = f'{dir}\\data\\arff\\{dataset_name}.arff'
    X, y, cat_len = load_arff_dataset(return_non_cat_length=True, path=path)
    print("\n\n Dataset ***", dataset_name, "*** number of not categorical categories: ", cat_len)
    X = pd.DataFrame(X)
    X['y'] = y
    enc = OneHotEncoder().fit(y.reshape(-1,1))
    print("Number of examples for each class:", np.sum(enc.transform(y.reshape(-1,1)).toarray(),0))
    display(X.describe(include='all'))
    display(X.sample(10))

    fig, axes = plt.subplots()
    sns.violinplot(data=pd.DataFrame(X), ax = axes)
    axes.yaxis.grid(True)
    axes.set_xlabel('Value')
    axes.set_ylabel('column')
    plt.show()
```
